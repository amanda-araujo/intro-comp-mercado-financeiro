## Criação de um portifólio

Lembrando da teoria... Da teoria moderna de portifólio (1959; Markowitz), temos que o **risco/volatilidade** de um portifólio **depende da correlação** entre os ativos.
*  Alta correlação: o risco dos ativos propaga-se para a carteira
*  Baixa correlação: menor o risco

Ao se criar um portifólio 50-50, o risco do portifólio é menor do que os riscos dos ativos individuais.

*Ideal*: busca por um alto número de ativos não correlacionados ao se montar uma carteira.

```python
# Criação de um portifólio
port_pesos = [0, 0, 0.5, 0, 0.5, 0]
dados['PORT1'] = dados.dot(port_pesos)
```

## Drawdown

Indicador, usado na prática ao se avaliar investimentos: Queda máxima de um ativo em relação ao ponto máximo anterior.
*   Drawdown = 0: ativo está em sua máxima histórica
*   != não está no máximo

```python
# Drawdown
ddown = pd.DataFrame()

for ativo in dados.columns:
  list = []
  for ind in range(dados.count()[0]): #calcula todas as diferenças
    list.append((dados[ativo].iloc[ind]/dados[ativo].iloc[:ind+1].max()-1) * 100)
  ddown[ativo] = list

ddown['Data'] = dados.index.values #vetor com histórico de drawdown
ddown.set_index(keys = 'Data', inplace = True)

print(ddown.min())
```

# **Fronteira Eficiente de Markowitz**

```python
# Eliminando o protifólio 1 dos dados
dados = dados.drop(['PORT1'], axis = 1)
dados_chg = dados_chg.drop(['PORT1'], axis = 1)
```

Definição de uma função para calcular o retorno e volatilidade de um portifólio. Passa-se uma lista de ativos, suas variações diárias e pesos para formar um portifólio.


```python
# Rentabilidade & volatilidade do portifólio

def calc_ret_vol(ativos, ativos_chg, port_pesos):
  port = ativos.dot(port_pesos)
  port_chg = port.pct_change()
  port_chg = port_chg.fillna(0)
  ret = ((port.iloc[-1]/port.iloc[0])**(1/5)) - 1 #retorno anualizado (5 anos)
  vol = port_chg.std()*np.sqrt(252)                #volatilidade anualizada
  return ret, vol
```

## Portifólio de 2 ativos

Queremos analisar várias ocmbinações de portifólio que incluem somente as duas ações GUAR3 e JBSS3. Cálculo de vários portifólios, variando de 5 em 5% a proporção dos ativos.

```python
# Portifólio com 2 ativos: JBS e GUAR
points = []
min_vol_ret = [100, 0] #[vol, ret]
min_port_pesos = [0, 0, 0, 0, 0, 0]

for w in range(0, 101, 5):
  port_pesos = [0, 0, w/100, 0, (1 - w/100), 0]
  ret, vol = calc_ret_vol(dados, dados_chg, port_pesos)

  print(f"Alocação: {round(w/100, 2):.2f} {round((1 - w/100), 2):.2f} Retorno: {round(ret, 3):.3f} Volatilidade: {round(vol, 3):.3f}")

  points.append([ret, vol]) #salvando dados

  # Estamos interessados no portifólio de menor volatilidade
  if (vol < min_vol_ret[0]):
    min_vol_ret[0] = vol
    min_vol_ret[1] = ret
    min_port_pesos[2] = w/100
    min_port_pesos[4] = 1 - w/100
```

```python
# Portifólio de menor volatilidade anualizada
print(min_vol_ret)
print(min_port_pesos)
```

> A combinação do portifólio com 75% de GUAR3 e 25% JBSS3 resulta na menor volatilidade.

```python
lp = np.array(points).T
plt.scatter(lp[[1][:]],lp[[0][:]]);
plt.ylabel("Retorno anualizado");
plt.xlabel("Volatilidade anualizada");
plt.title("Batirangue")

plt.scatter(vol_aa['GUAR3.SA'], ret_aa['GUAR3.SA'], color='red');
plt.text(vol_aa['GUAR3.SA'], ret_aa['GUAR3.SA'], 'GUAR3');

plt.scatter(vol_aa['JBSS3.SA'], ret_aa['JBSS3.SA'], color='red');
plt.text(vol_aa['JBSS3.SA'], ret_aa['JBSS3.SA'], 'JBSS3');

plt.scatter(min_vol_ret[0], min_vol_ret[1], color='green');
plt.text(min_vol_ret[0], min_vol_ret[1], 'Min. Vol.');
```

> Testamos várias combinações de portifólio (força bruta) e dentre elas pegamos a de menor volatilidade. Construção da hipérbole, a fronteira eficiente de Markowitz: pontos entre a carteira de variância mínima até o topo do braço da hipérbole (que termina em 100% de uma das ações, no caso GUAR3). 

Vamos criar um portifólio usando a porporção de menor volatilidade.

```python
dados['PORT_min_vol'] = dados.dot(min_port_pesos)
dados_chg = dados.pct_change()
dados_chg = dados_chg.fillna(0)
```

## Criando portifólio com 3 ativos

```python
# Eliminando o portifólio anterior dos dados
dados = dados.drop(['PORT_min_vol'], axis = 1)
dados_chg = dados_chg.drop(['PORT_min_vol'], axis = 1)
```

```python
# Portifólio com 3 ativos: JBSS3, GUAR3 e PSSA3
points = []
min_vol_ret = [100, 0] #[vol, ret]
min_port_pesos = [0, 0, 0, 0, 0, 0]

for w1 in range(0, 101, 5): #ativo 1
  for w2 in range(0, 101 - w1, 5): #ativo 2

    port_pesos = [0, 0, w1/100, 0, w2/100, (1 - w1/100 - w2/100)]
    ret, vol = calc_ret_vol(dados, dados_chg, port_pesos)

    print(f"Alocação: {round(w1/100, 2):.2f} {round(w2/100, 2):.2f} {round((1 - w1/100 - w2/100), 2):.2f} Retorno: {round(ret, 3):.3f} Volatilidade: {round(vol, 3):.3f}")

    points.append([ret, vol]) #salvando dados

    # Estamos interessados no portifólio de menor volatilidade
    if (vol < min_vol_ret[0]):
      min_vol_ret[0] = vol
      min_vol_ret[1] = ret
      min_port_pesos[2] = w1/100
      min_port_pesos[4] = w2/100
      min_port_pesos[5] = 1 - w1/100 - w2/100
```
